```python
import os
from campaign_granule import CampaignGranule
from converters.point_cloud_converter import PointCloudConverter
from zarr_writer.point_cloud_zarr import PointCloudToZarrConverter
from zarr_writer.zarr_writer import ZarrWriter

# Path to a small test NetCDF file
fcx_dir = os.path.expanduser(r'~\Documents\GHRC\FCX')
nc_filename = 'GOESR_CRS_L1B_20170517_v0.nc'
nc_filepath = os.path.join(fcx_dir, nc_filename)
netcdf_path = nc_filepath
output_dir = os.path.join(fcx_dir, "output")
output_path = os.path.join(output_dir, 'zarr_test_output')
os.makedirs(output_path, exist_ok=True)

# Load granule and convert to point cloud
print("Loading granule.")
granule = CampaignGranule(netcdf_path)
print("Converting to PointCloud.")
cloud = PointCloudConverter().convert(granule)

# Ensure output folder exists
os.makedirs(output_path, exist_ok=True)

# Write to Zarr
print("Writing to Zarr.")
writer = ZarrWriter(output_path, strategy=PointCloudToZarrConverter())
writer.write(cloud)

print(f"✅ Zarr written to {output_path}")
```

```python
# Verify output

import xarray as xr

ds = xr.open_zarr(output_path, consolidated=False)

#print(ds)
#print(ds["location"].shape)
print(ds["time"].values[:5])
#print(ds["ref"].values[:5])
```

```python
import xarray as xr

def inspect_group(group, prefix=""):
    for name, array in group.arrays():
        path = f"{prefix}/{name}" if prefix else name
        print(f"Inspecting {path}...")
        print("  dimension_names:", array.attrs.get("dimension_names", "❌ missing"))
    for name, subgroup in group.groups():
        inspect_group(subgroup, prefix=f"{prefix}/{name}" if prefix else name)

try:
    #ds = xr.open_zarr("tests/output/zarr_test_output", consolidated=False)
    ds = xr.open_zarr(output_path, group=None, consolidated=False)

except KeyError as e:
    print("KeyError:", e)
    import zarr
    root = zarr.open(output_path, mode="r")
    inspect_group(root)
    #print(root["location"][:5])

```
